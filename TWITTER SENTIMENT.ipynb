```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet 
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
```

```python
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ # adjective 
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV # adverb 
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords 
import string
stops = set(stopwords.words('english'))

punct = list(string.punctuation)
stops.update(punct)
```

```python
def clean_review(words):
    output_words = [] 
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w]) # POS of original words as lowercase will loose some info 
            
            # getting the tuple and then the pos tag as shown above with 'better'
            clean_word = lemmatizer.lemmatize(w , pos = get_simple_pos(pos[0][1])) 
            output_words.append(clean_word.lower())
    return output_words   
```

```python
documents = [(clean_review(document) , category) for document ,  category in documents]
```

```python
training_documents = documents[:7000] 
testing_documents = documents[7000:]
```

```python
training_documents_orig = documents
testing_documents_orig = test_documents
```

```python
all_words = [] 

for doc in training_documents:
    all_words += doc[0]
```

```python
all_words_orig = []
for doc in training_documents_orig:
    all_words_orig += doc[0]
```

```python
import nltk 
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
freq = nltk.FreqDist(all_words_orig)
common = freq.most_common(3000)
features_orig = [i[0] for i in common]
```

```python
def get_feature_dict(words):
    current_features = {} 
    words_set = set(words)
    
    # for each word in feature we need to check if it comes in the doc or not 
    for w in features: 
        current_features[w]= w in words_set # is true if word in words set
    return current_features
```

```python
output = get_feature_dict(training_documents[0][0])
output
```

```python
training_data = [(get_feature_dict(doc) , category) for doc , category in training_documents]
```

```python
testing_data = [(get_feature_dict(doc) , category) for doc , category in testing_documents]
```

```python
training_data_orig = [(get_feature_dict(doc), category) for doc, category in training_documents_orig]
testing_data_orig = [(get_feature_dict(doc)) for doc in testing_documents_orig]
```

```python
from sklearn.svm import SVC
from nltk.classify.scikitlearn import SklearnClassifier
```

```python
svc = SVC()
classifier_sklearn = SklearnClassifier(svc)
```

```python
classifier_sklearn.train(training_data)
```

```python
nltk.classify.accuracy(classifier_sklearn , testing_data)
```
